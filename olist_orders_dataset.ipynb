```python
date_cols = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]

for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')
```

Handling Missing Values Carefully......

```python
df[['order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date']].isnull().sum()
```

```python
# 1ï¸âƒ£ Create derived columns
# Approval time: difference between approval and purchase
df['approval_time_hours'] = (df['order_approved_at'] - df['order_purchase_timestamp']).dt.total_seconds() / 3600

# Delivery time from carrier: difference between carrier delivery and purchase
df['carrier_delivery_time_hours'] = (df['order_delivered_carrier_date'] - df['order_purchase_timestamp']).dt.total_seconds() / 3600

# Delivery time to customer: difference between customer delivery and purchase
df['customer_delivery_time_hours'] = (df['order_delivered_customer_date'] - df['order_purchase_timestamp']).dt.total_seconds() / 3600

```

```python
print(df[['approval_time_hours', 'carrier_delivery_time_hours', 'customer_delivery_time_hours',]].head())
```

```python
#here , some tieme values are negative which is not possible so we will replace them with NaN
# Replace negative values with NaN
time_cols = ['approval_time_hours', 'carrier_delivery_time_hours', 
             'customer_delivery_time_hours']

for col in time_cols:
    df[col] = df[col].apply(lambda x: x if x >= 0 else None)
```

```python
print(df[['approval_time_hours', 'carrier_delivery_time_hours', 'customer_delivery_time_hours']].head())
```
