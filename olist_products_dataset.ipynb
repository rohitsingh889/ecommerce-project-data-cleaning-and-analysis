```python
# Drop rows with missing values in high-impact columns only
high_impact_cols = [
    'product_weight_g',
    'product_length_cm',
    'product_height_cm',
    'product_width_cm'
]

df.dropna(subset=high_impact_cols, inplace=True)

# Optional: Confirm the result
print(f"✅ Remaining rows after drop: {df.shape[0]}")
print(df.isna().sum())
```

```python
# Fill with median values for numerical columns
df['product_name_lenght'].fillna(df['product_name_lenght'].median(), inplace=True)
df['product_description_lenght'].fillna(df['product_description_lenght'].median(), inplace=True)

# Fill with default value for photo count
df['product_photos_qty'].fillna(1, inplace=True)  # Assuming 1 photo is a safe default
```
